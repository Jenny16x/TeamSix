```python
# import dependencies
import pandas as pd
from path import Path
```

```python
# read in csv file
# Read in the data
data = Path('LahmanData/Pitching.csv')
pitch_df = pd.read_csv(data)
pitch_df
```

```python
# check datatypes of columns
pitch_df.dtypes
```

```python
# filter to include only years 2016-2019
pitch_yrs_df = pitch_df[(pitch_df.yearID>2015) & (pitch_df.yearID<2020)]
pitch_yrs_df
```

```python
# export pitcher stats as csv
pitch_yrs_df.to_csv('2016-2019_pitcher_stats.csv', index=True)
```

```python
# read in csv file
# Read in the data
salary_data = Path('LahmanData/Salaries.csv')
salary_df = pd.read_csv(salary_data)
salary_df
```

```python
# filter to include only years 2016-2019
salary_df_filt = salary_df[(salary_df.yearID>2015) & (salary_df.yearID<2020)]
salary_df_filt
```

```python
# merge pitcher stats + salary data
stats_salaries_df=pitch_yrs_df.merge(salary_df_filt, how='left', on='playerID')
stats_salaries_df
```

```python
# try inner join
# merge pitcher stats + salary data
stats_salaries_inner=pitch_yrs_df.merge(salary_df_filt, how='inner', on='playerID')
stats_salaries_inner
```

```python
# drop duplicate columns from y-table (salaries csv)
statSal_df=stats_salaries_df.drop(['yearID_y','teamID_y','lgID_y'], axis=1)
statSal_df
```

```python
# export pitcher stats as csv
stats_salaries_df.to_csv('2016-2019_pitcher_stats_n_sal.csv', index=False)
```

```python
# look at column names
pitch_yrs_df.columns
```

```python
# list of all playerIDs included
pitch_yrs_df.playerID
```

```python
# only include unique playerIDs (removes duplicates from mult years)
unique_IDs=pitch_yrs_df.playerID.unique()
unique_IDs
```

```python
# get amount of unique playerIDs
len(unique_IDs)
```

```python
# convert unique playerIDs array into dataframe
unique_df=pd.DataFrame(unique_IDs,columns=['playerID'])
unique_df
```

```python
# import people data
ppl_data = Path('LahmanData/People.csv')
ppl_df = pd.read_csv(ppl_data)
ppl_df
```

```python
# merge unique pitcher playerIDs x People csv with player info
pitch_ppl_df=unique_df.merge(ppl_df, how='left', on='playerID')
pitch_ppl_df
```

```python
# look at columns in new df
pitch_ppl_df.columns
```

```python
# remove unnecessary columns from dataframe
pitchDem_df=pitch_ppl_df.drop(['bats','birthCountry','birthState','birthCity','birthDay','birthCountry','birthState','birthCity','deathYear','deathMonth','deathDay','deathCountry','deathState','deathCity','finalGame'], axis=1)
pitchDem_df
```

```python
# export as pitchDem_df as csv
pitchDem_df.to_csv('pitch-demographics.csv', index=True)
```
