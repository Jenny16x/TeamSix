```python
# import dependencies
# import dependencies
import pandas as pd
from path import Path
```

```python
# first import pitching stats data
data = Path('../LahmanData/Pitching.csv')
pitch_df = pd.read_csv(data, index_col=False)
pitch_df
```

```python
# filter to include years 2010-2019
yrs_pitch_df = pitch_df[(pitch_df.yearID>2009) & (pitch_df.yearID<2020)]
yrs_pitch_df
```

```python
# drop NaN values
yrs_pitch_df=yrs_pitch_df.dropna()
```

```python
# combine playerID and yearID to create ID
yrs_pitch_df["ID"] = yrs_pitch_df["yearID"].map(str) + yrs_pitch_df["playerID"]
yrs_pitch_df
```

```python
# # drop playerID and yearID columns
# pitch_df_clean=yrs_pitch_df.drop(['playerID','yearID'], axis=1)
# pitch_df_clean
```

```python
# shift ID column to beginning
first_col = yrs_pitch_df.pop('ID')
```

```python
# insert column to new position
yrs_pitch_df.insert(0,'ID',first_col)
yrs_pitch_df
```

```python
# export cleaned pitching stats as csv
yrs_pitch_df.to_csv('2010-2019_pitching_clean.csv', index=False)
```
