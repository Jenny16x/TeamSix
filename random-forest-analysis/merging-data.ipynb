```python
# import dependencies
import pandas as pd
from path import Path
```

```python
# import cleaned pitching data
data = Path('2010-2019_pitching_clean.csv')
pitch_df = pd.read_csv(data, index_col=False)
pitch_df
```

```python
# import cleaned people data
data = Path('2010-2019_people_clean.csv')
ppl_df = pd.read_csv(data, index_col=False)
ppl_df
```

```python
pitch_ppl = pd.merge(left=pitch_df, right=ppl_df, how='left', left_on='playerID', right_on='playerID')
pitch_ppl
```

```python
# see if people data merged properly
pd.set_option('display.max_columns', None)
pitch_ppl.loc[pitch_ppl['playerID'] == 'abadfe01']
```

```python
# # drop playerID, yearID cols from pitch_ppl df to prep for merge with sal_df
pitch_ppl=pitch_ppl.drop(['playerID','yearID'], axis=1)
pitch_ppl
```

```python
# import cleaned salaries data
# first import pitching stats data
data = Path('2010-2019_salaries_clean.csv')
sal_df = pd.read_csv(data, index_col=False)
sal_df=sal_df.dropna()
sal_df
```

```python
sal_df.loc[sal_df['playerID']=='abadfe01']
```

```python
# make sure no nan vals
sal_df.isnull().any().any()
```

```python
# # drop playerID, yearID, teamID, lgID cols from sal_df to prep for merge
sal_df=sal_df.drop(['playerID','yearID','teamID','lgID'], axis=1)
sal_df
```

```python
# left join pitch_ppl + sal_df
pitch_ppl_sal = pd.merge(left=pitch_ppl, right=sal_df, how='left', left_on='ID', right_on='ID')
pitch_ppl_sal
```

```python
# drop nan values
pitch_ppl_sal=pitch_ppl_sal.dropna()
pitch_ppl_sal
```

```python
# export dataset as csv
# export cleaned pitching stats as csv
pitch_ppl_sal.to_csv('2010-2019_data.csv', index=False)
```
