```python
import pandas as pd
from sklearn.decomposition import PCA
import numpy
import matplotlib.pyplot as plot
from path import Path
```

```python
# Loading data
file_path = Path("encoded_pitcher_stat_sal_dem.csv")
df = pd.read_csv(file_path)
df
```

```python
df['salary'].describe()
```

```python
sal_bins=pd.qcut(df['salary'], q=3)
```

```python
sal_bins.value_counts()
```

```python
# Adding Labels to Pandas .qcut()
df['salary-bin'] = pd.qcut(
   df['salary'], 
   q=3, 
   labels=['low', 'mid', 'high']
)
print(df.head())
```
