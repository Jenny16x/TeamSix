```python
# STRIKE OUTS
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Resources/pitcher_salaries_cleaned.csv")
X = df["Strike Outs"].array.reshape(-1, 1)
y = df["Salary"].values

#Fitting Linear Regression
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
plt.scatter(X,y, color="red")
plt.plot(X, lin_reg.predict(X))
plt.title("Linear Regression")
plt.xlabel("Strike Outs")
plt.ylabel("Salary")
plt.show()
```

```python
#Linear Regression prediction Test
lin_reg.predict([[100]])
```

```python
#Linear Regression prediction Test
lin_reg.predict([[300]])
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
print((y_test, y_pred))
```

```python
# DEGREE OF 2

poly = PolynomialFeatures(degree=2, include_bias=True)
x_train_trans = poly.fit_transform(x_train)
x_test_trans = poly.transform(x_test)

#include bias parameter
lr = LinearRegression()
lr.fit(x_train_trans, y_train)
y_pred = lr.predict(x_test_trans)

import numpy as np
X_new = np.linspace(0, 10, 400).reshape(-1, 1)
X_new_poly = poly.transform(X_new)
y_new = lr.predict(X_new_poly)
plt.plot(X_new, y_new, "r-", linewidth=2, label="Predictions")
plt.plot(x_train, y_train, "b.",label='Training points')
plt.plot(x_test, y_test, "g.",label='Testing points')
plt.xlabel("ERA")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

```python
# Polynomial Regression prediction
new_salary_pred = y_new[100]
print('The predicted salary for a pitcher with 100 Strike Outs 1 is ',new_salary_pred)

# Polynomial Regression prediction
new_salary_pred = y_new[300]
print('The predicted salary for a pitcher with 300 Strike Outs is ',new_salary_pred)
```

```python
# DEGREE OF 5

poly = PolynomialFeatures(degree=5, include_bias=True)
x_train_trans = poly.fit_transform(x_train)
x_test_trans = poly.transform(x_test)

#include bias parameter
lr = LinearRegression()
lr.fit(x_train_trans, y_train)
y_pred = lr.predict(x_test_trans)

import numpy as np
X_new = np.linspace(0, 10, 400).reshape(-1, 1)
X_new_poly = poly.transform(X_new)
y_new = lr.predict(X_new_poly)
plt.plot(X_new, y_new, "r-", linewidth=2, label="Predictions")
plt.plot(x_train, y_train, "b.",label='Training points')
plt.plot(x_test, y_test, "g.",label='Testing points')
plt.xlabel("ERA")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

```python
# Polynomial Regression prediction
new_salary_pred = y_new[100]
print('The predicted salary for a pitcher with 100 Strike Outs 1 is ',new_salary_pred)

# Polynomial Regression prediction
new_salary_pred = y_new[300]
print('The predicted salary for a pitcher with 300 Strike Outs is ',new_salary_pred)
```

```python
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X)
```

```python
# Passing X_poly to LinearRegression
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y)

# Polynomial Regression prediction
new_salary_pred = lin_reg_2.predict(poly_reg.fit_transform([[100]]))
print('The predicted salary for a pitcher who threw 100 Strike Outs is ',new_salary_pred)

# Polynomial Regression prediction
new_salary_pred = lin_reg_2.predict(poly_reg.fit_transform([[300]]))
print('The predicted salary for a pitcher who threw 300 Strike Outs is ',new_salary_pred)
```

```python
# Visualize Poly Regression Results
plt.scatter(X,y, color="red")
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)))
plt.title("Poly Regression - Degree 2")
plt.xlabel("Strike Outs")
plt.ylabel("Salary")
plt.show()
```

```python
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)

# Passing X_poly to LinearRegression
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y)

# Visualize Poly Regression Results
plt.scatter(X,y, color="red")
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)))
plt.title("Poly Regression - Degree 4")
plt.xlabel("Strike Outs")
plt.ylabel("Salary")
plt.show()
```

```python
# Polynomial Regression prediction
new_salary_pred = lin_reg_2.predict(poly_reg.fit_transform([[100]]))
print('The predicted salary for a pitcher who threw 100 Strike Outs is ',new_salary_pred)

# Polynomial Regression prediction
new_salary_pred = lin_reg_2.predict(poly_reg.fit_transform([[300]]))
print('The predicted salary for a pitcher who threw 300 Strike Outs is ',new_salary_pred)
```
