```python
# import dependencies
import pandas as pd
from path import Path
```

```python
# import cleaned pitching data
data = Path('encoded_data.csv')
df = pd.read_csv(data, index_col=False)
df
```

```python
# look at salary stats distribution
df['salary'].describe()
```

```python
# create bins for salary
sal_bins=pd.qcut(df['salary'], q=3)
```

```python
# see how many rows in each bin
sal_bins.value_counts()
```

```python
# Adding column + labels for salary bins
df['salary-bin'] = pd.qcut(
   df['salary'], 
   q=3, 
   labels=['low', 'mid', 'high']
)
df.head()
```

```python
# export as csv
df.to_csv('encoded_data_binned.csv', index=False)
```
