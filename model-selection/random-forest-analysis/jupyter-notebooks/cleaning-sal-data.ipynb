```python
# import dependencies
# import dependencies
import pandas as pd
from path import Path
```

```python
# first import salary stats data
data = Path('./LahmanData/Salaries.csv')
sal_df = pd.read_csv(data, index_col=False)
sal_df
```

```python
# drop NaN values
sal_df=sal_df.dropna()
sal_df
```

```python
# filter to include years 2010-2019
sal_df = sal_df[(sal_df.yearID>2009) & (sal_df.yearID<2020)]
sal_df
```

```python
# combine playerID and yearID to create ID
sal_df["ID"] = sal_df["yearID"].map(str) + sal_df["playerID"]
sal_df
```

```python
# shift ID column to beginning
first_col = sal_df.pop('ID')
```

```python
# insert column to new position
sal_df.insert(0,'ID',first_col)
sal_df
```

```python
# export cleaned pitching stats as csv
sal_df.to_csv('2010-2019_salaries_clean.csv', index=False)
```
