```python
#Load Data 
# Use ERA for 1st Variable Test
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures

df = pd.read_csv("Resources/pitcher_salaries_cleaned.csv")
X = df["ERA"].array.reshape(-1, 1)
y = df["Salary"].values
```

```python
#Fitting Linear Regression
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
# Visualize Linear Regression Results
import matplotlib.pyplot as plt

plt.scatter(X,y, color="red")
plt.plot(X, lin_reg.predict(X))
plt.title("Linear Regression")
plt.xlabel("ERA")
plt.ylabel("Salary")
# plt.ylim(0.0, 30.0)
# plt.xlim(0.0, 20.0)
plt.show()
```

```python
#Linear Regression prediction Test
lin_reg.predict([[1]])
```

```python
# Linear Regression prediction Test
lin_reg.predict([[10]])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
print((y_test, y_pred))
```

```python
plt.plot(x_train, lr.predict(x_train), color="r")
plt.scatter(X,y, color="blue")
plt.xlabel("ERA")
plt.ylabel("Salary")
plt.show()
```

```python
# DEGREE OF 2

poly = PolynomialFeatures(degree=2, include_bias=True)
x_train_trans = poly.fit_transform(x_train)
x_test_trans = poly.transform(x_test)

#include bias parameter
lr = LinearRegression()
lr.fit(x_train_trans, y_train)
y_pred = lr.predict(x_test_trans)

import numpy as np
X_new = np.linspace(0, 10, 200).reshape(-1, 1)
X_new_poly = poly.transform(X_new)
y_new = lr.predict(X_new_poly)
plt.plot(X_new, y_new, "r-", linewidth=2, label="Predictions")
plt.plot(x_train, y_train, "b.",label='Training points')
plt.plot(x_test, y_test, "g.",label='Testing points')
plt.xlabel("ERA")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

```python
# DEGREE OF 2
# Polynomial Regression prediction
new_salary_pred = y_new[1]
print('The predicted salary for a pitcher with an ERA of 1 is ',new_salary_pred)

# Polynomial Regression prediction
new_salary_pred = y_new[10]
print('The predicted salary for a pitcher with an ERA of 10 is ',new_salary_pred)
```
