```python
# import dependencies
import pandas as pd
from path import Path
```

```python
# import cleaned pitching data
data = Path('2010-2019_data.csv')
df = pd.read_csv(data, index_col=False)
df
```

```python
df.dtypes
```

```python
# encode object features
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
encoded_df = df.copy()
df['teamID'] = le.fit_transform(df['teamID'])
df['lgID'] = le.fit_transform(df['lgID'])
df['throws'] = le.fit_transform(df['throws'])
df['bats'] = le.fit_transform(df['bats'])

df
```

```python
# export as full dataframe as csv
df.to_csv('encoded_data', index=False)
```
