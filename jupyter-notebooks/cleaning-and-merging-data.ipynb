```python
# import dependencies
import pandas as pd
from path import Path
```

```python
# read in csv file
# Read in the data
data = Path('2016-2019_pitcher_stats.csv')
pitcher_df = pd.read_csv(data, index_col=False)
pitcher_df
```

```python
# read in csv file
# Read in the data
salary_data = Path('LahmanData/Salaries.csv')
salary_df = pd.read_csv(salary_data)
# Drop the null columns where all values are null
salary_df = salary_df.dropna(axis='columns', how='all')

# Drop the null rows
salary_df = salary_df.dropna()

salary_df
```

```python
# filter to include only years 2016-2019
filt_sal_df = salary_df[(salary_df.yearID>2015) & (salary_df.yearID<2020)]
filt_sal_df
```

```python
# merge pitcher stats with pitcher salaries
merged_df = pd.merge(pitcher_df, filt_sal_df,  how='left', left_on=['playerID','yearID','teamID','lgID'], right_on = ['playerID','yearID','teamID','lgID'])
merged_df=merged_df.dropna()
merged_dfk9
```

```python
# read in csv file
# Read in the data
dem_data = Path('junk/pitch-demographics.csv')
dem_df = pd.read_csv(dem_data, index_col=False)
dem_df=dem_df.drop(['nameFirst','nameLast','retroID','bbrefID','nameGiven'], axis=1)
dem_df
```

```python
# merge demographics info with pitcher stat+sal df
full_df=merged_df.merge(dem_df, how='left', on='playerID')
full_df=full_df.dropna()
full_df
```
