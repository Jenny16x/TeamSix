```python
import pandas as pd
```

```python
# Files to load
file_to_load = "Resources/Pitching.csv"

# Read the CSV into a DataFrame
pitching = pd.read_csv(file_to_load)
pitching
```

```python
# Files to load
file_to_load = "Resources/Salaries.csv"

# Read the CSV into a DataFrame
salaries = pd.read_csv(file_to_load)
salaries
```

```python
# Merge pitchers and salaries files
pitcher_salary = pd.merge(pitching, salaries, how='left')
pitcher_salary
```

```python
pitcher_salary.dtypes
```

```python
# Filter data by years after 1990
pitcher_salary2k = pitcher_salary[pitcher_salary['yearID']>=1990]
pitcher_salary2k
```

```python
# Filter data by years before 2020 to avoid possible Coivd data skew 
pitcher_salary_years = pitcher_salary2k[pitcher_salary2k['yearID']<2020]
pitcher_salary_years
```

```python
# Filter data by pitchers who have started more than 3 games in a year 
pitcher_salary_gs = pitcher_salary_years[pitcher_salary_years['GS']>=3]
pitcher_salary_gs
```

```python
# Drop rows without salary data
psfull = pitcher_salary_gs.dropna(subset=["salary"])
psfull
```

```python
list(psfull.columns)
```

```python
# Drop columns not relevent 
ps_reduced = psfull.drop(["SV", "stint", "BAOpp", "IBB", "HBP", "BK", "SH", "SF", "GIDP", "CG", "SHO", "BFP", "GF", "R", "WP", "IPouts", "BB"], 1)
ps_reduced
```

```python
list(ps_reduced.columns)
```

```python
# loading in  People csv
people = pd.read_csv("Resources/People.csv")
people.head()
```

```python
people.columns
```

```python
people = people[["playerID", "birthYear", "nameFirst", "nameLast", "weight", "height"]]
```

```python
people.head()
```

```python
# Merge people and ps_reduced files
ps_info = pd.merge(ps_reduced, people, how='left')
ps_info
```

```python
# Change columns names to full names
ps_info = ps_info.rename(columns={"salary" : "Salary",
                                  "yearID" : "Year",
                                "teamID" : "Team",
                                  "lgID" : "League",
                                  "H" : "Hits",
                                  "W" : "Wins",
                                  "L" : "Losses",
                                  "G" : "Games",
                                  "GS" : "Games Started",
                                  "ER" : "Earned Runs",
                                  "HR" : "Home Runs",
                                  "SO" : "Strike Outs",
                                  "nameFirst" : "First Name",
                                  "nameLast" : "Last Name",
                                  "birthYear" : "Year Born",
                                  "weight" : "Weight",
                                  "height" : "Height",
                                 })
ps_info
```

```python
ps_info.columns
```

```python
# Rearrange Columns
ps_info = ps_info[["playerID", "Year", "Last Name", "First Name", "Salary", "ERA", "Hits", "Earned Runs","Strike Outs", "Home Runs", "Games Started", "Wins", "Losses", "Games", "Weight", "Height","Year Born", "League", "Team" ]]
ps_info
```

```python
# Calculate Age using Year less Year Born
ps_info["Age"] = ps_info["Year"] - ps_info["Year Born"]
ps_info
```

```python
# Rearrange Columns
ps_info = ps_info[["playerID", "Year", "Last Name", "First Name", "Age", "Salary", "ERA", "Hits", "Earned Runs","Strike Outs", "Home Runs", "Wins", "Losses", "Weight", "Height", "League", "Team", "Games Started" ]]
ps_info
```

```python
# Combine Names
ps_info["Full Name"] = ps_info["Last Name"] + ps_info["First Name"]
ps_info
```

```python
# Rearrange Columns
ps_info = ps_info[["playerID", "Year", "Full Name", "Age", "Salary", "ERA", "Hits", "Earned Runs","Strike Outs", "Home Runs", "Wins", "Losses", "Weight", "Height", "League", "Team", "Games Started" ]]
ps_info
```
